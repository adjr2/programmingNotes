```python
# Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating the DataFrame

dataFull = pd.DataFrame({
    'Question': ['Quality of products', 'Do you like our shop?', 'Customer service', 'Overall experience'],
    'excellent': [191, 448, 63, 40],
    'good': [184, 16, 127, 100],
    'average': [133, 12, 171, 126],
    'poor': [82, 146, 191, 52],
    'appalling': [25, 0, 58, 26],
})
```

```python
fig, axs = plt.subplots(2, 2)

fig.set_figheight(10)
fig.set_figwidth(10)

axs[0, 0].bar(dataFull.columns[1:], dataFull.loc[0][1:]);
axs[0, 0].set_title(dataFull.loc[0][0])

for i in range(len(dataFull.columns[1:])):
    axs[0, 0].text(i, dataFull.loc[0][1:][i], dataFull.loc[0][1:][i], ha = 'center')

axs[0, 0].set_xlabel("Rating Scale")
axs[0, 0].set_ylabel("Percentage of Responses")


axs[1, 0].bar(dataFull.columns[1:], dataFull.loc[1][1:]);
axs[1, 0].set_title(dataFull.loc[1][0])

for i in range(len(dataFull.columns[1:])):
    axs[1, 0].text(i, dataFull.loc[1][1:][i], dataFull.loc[1][1:][i], ha = 'center')

axs[1, 0].set_xlabel("Rating Scale")
axs[1, 0].set_ylabel("Percentage of Responses")



axs[0, 1].bar(dataFull.columns[1:], dataFull.loc[2][1:]);
axs[0, 1].set_title(dataFull.loc[2][0])

for i in range(len(dataFull.columns[1:])):
    axs[0, 1].text(i, dataFull.loc[2][1:][i], dataFull.loc[2][1:][i], ha = 'center')

axs[0, 1].set_xlabel("Rating Scale")
axs[0, 1].set_ylabel("Percentage of Responses")


axs[1, 1].bar(dataFull.columns[1:], dataFull.loc[3][1:]);
axs[1, 1].set_title(dataFull.loc[3][0])

for i in range(len(dataFull.columns[1:])):
    axs[1, 1].text(i, dataFull.loc[3][1:][i], dataFull.loc[3][1:][i], ha = 'center')

axs[1, 1].set_xlabel("Rating Scale")
axs[1, 1].set_ylabel("Percentage of Responses")

fig.suptitle('Customer Survey', size=18)
fig.tight_layout()
```

#### Normalizing the data
- As `not all submitted questionnaires were necessarily complete`, we decided to calculate the percentage of responses for each survey question. 

```python
# Calculating the percentage
dataAverage = dataFull.copy()

# adding total number of responses
total = []
for i in range(dataAverage.shape[0]):
    total.append(dataAverage.loc[i][1:].sum())
dataAverage['total'] = total

for i in range(dataAverage.shape[0]):
    a = np.array((dataAverage.loc[i,dataAverage.columns[1:-1]]*100)/total[i])
    dataAverage.loc[i,dataAverage.columns[1:-1]] = np.round(a.astype(np.double),0)
```

```python
fig, axs = plt.subplots(2, 2)

fig.set_figheight(10)
fig.set_figwidth(10)


axs[0, 0].bar(dataAverage.columns[1:-1], dataAverage.loc[0][1:-1]);
axs[0, 0].set_title(dataFull.loc[0][0])

for i in range(len(dataAverage.columns[1:-1])):
    axs[0, 0].text(i, dataAverage.loc[0][1:-1][i], dataAverage.loc[0][1:-1][i], ha = 'center')

axs[0, 0].set_xlabel("Rating Scale")
axs[0, 0].set_ylabel("Percentage of Responses")
 

axs[1, 0].bar(dataAverage.columns[1:-1], dataAverage.loc[1][1:-1]);
axs[1, 0].set_title(dataAverage.loc[1][0])

for i in range(len(dataAverage.columns[1:-1])):
    axs[1, 0].text(i, dataAverage.loc[1][1:-1][i], dataAverage.loc[1][1:-1][i], ha = 'center')

axs[1, 0].set_xlabel("Rating Scale")
axs[1, 0].set_ylabel("Percentage of Responses")


axs[0, 1].bar(dataAverage.columns[1:-1], dataAverage.loc[2][1:-1]);
axs[0, 1].set_title(dataFull.loc[2][0])

for i in range(len(dataAverage.columns[1:-1])):
    axs[0, 1].text(i, dataAverage.loc[2][1:-1][i], dataAverage.loc[2][1:-1][i], ha = 'center')
axs[0, 1].set_xlabel("Rating Scale")
axs[0, 1].set_ylabel("Percentage of Responses")

axs[1, 1].bar(dataAverage.columns[1:-1], dataAverage.loc[3][1:-1]);
axs[1, 1].set_title(dataAverage.loc[3][0])

for i in range(len(dataAverage.columns[1:-1])):
    axs[1, 1].text(i, dataAverage.loc[3][1:-1][i], dataAverage.loc[3][1:-1][i], ha = 'center')
axs[1, 1].set_xlabel("Rating Scale")
axs[1, 1].set_ylabel("Percentage of Responses")

fig.suptitle('Customer Survey (Percentage)', size=18)
fig.tight_layout()
```

```python
fig, axs = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(10)

bottom = [0,0,0,0]
axs.bar(dataFull.loc[:,dataFull.columns[0]], dataFull.loc[:,dataFull.columns[1]], bottom=bottom)

bottom += dataFull.loc[:,dataFull.columns[1]]
axs.bar(dataFull.loc[:,dataFull.columns[0]], dataFull.loc[:,dataFull.columns[2]], bottom=bottom)

bottom += dataFull.loc[:,dataFull.columns[2]] 
axs.bar(dataFull.loc[:,dataFull.columns[0]], dataFull.loc[:,dataFull.columns[3]], bottom=bottom)

bottom += dataFull.loc[:,dataFull.columns[3]]
axs.bar(dataFull.loc[:,dataFull.columns[0]], dataFull.loc[:,dataFull.columns[4]], bottom=bottom)

bottom += dataFull.loc[:,dataFull.columns[4]]
axs.bar(dataFull.loc[:,dataFull.columns[0]], dataFull.loc[:,dataFull.columns[5]], bottom=bottom)

for c in axs.containers:
    labels = [v.get_height() if v.get_height() > 0 else '' for v in c]
    axs.bar_label(c, labels=labels, label_type='center')

axs.set_xlabel("Survey Questions")
axs.set_ylabel("Number of Responses")
axs.legend(["excellent", "good", "average", "poor", "appalling"])
axs.set_title("Customer Survey", size=18);
# axs.show()
```
